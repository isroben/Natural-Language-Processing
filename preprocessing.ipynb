## URL's removal

```python
# defining a function
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

```python
text1 = 'check out my github https://www.github.com/isroben'
text2 = 'check out my linkedin http://www.linkedin.com/isroben'
```

```python
remove_url(text1), remove_url(text2)
```

## Removing Punctuation

### Hard Coding method

```python
exclude = st.punctuation
```

```python
def remove_punc0(text):
    for char in exclude:
        text = text.replace(char, '')
    return text
```

```python
text = 'string.with.punc?'
```

```python
start = time.time()
data['review'].apply(remove_punc0)
time1 = time.time() - start
time1
```

### Standard and Fast Method

```python
def remove_punc1(text):
    return text.translate(str.maketrans('', '',exclude))
```

```python
start = time.time()
data['review'].apply(remove_punc1)
time1 = time.time() - start
time1
```

## Stop Words removal

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
def remove_stopwords(text):
    return " ".join(
        [word for word in text.split() if word.lower() not in stop_words]
    )
```

```python
data['review'] = data['review'].apply(lambda x: remove_stopwords(x) if isinstance(x, str) else x)
```

## Stemming

It is used in Information Retrival system

```python
from nltk import PorterStemmer
```

```python
ps = PorterStemmer() 
def stem_words(text):
    return " ".join(([ps.stem(word) for word in text.split()]))
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

```python
text = "probably my alltime favorite movie a story of selflessness sacrifice and dedication to a noble cause but its not prechy or boring it just never gets old"
```

```python
stem_words(text)
```

## Lemmatization

```python
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = 'He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun.'

punctuations = '?:!.,;'
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
print ("{0:20}{1:20}".format("Word", "Lemma"))
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word, pos='v')))
```
