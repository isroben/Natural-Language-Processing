## Removing tags and notations

```python
import re
```

```python
def remove_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

```python
data['review'] = data['review'].apply(remove_tags)
```

## URL's removal

```python
# defining a function
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

```python
text1 = 'check out my github https://www.github.com/isroben'
text2 = 'check out my linkedin http://www.linkedin.com/isroben'
```

```python
remove_url(text1), remove_url(text2)
```

## Removing Punctuation

### Hard Coding method

```python
exclude = st.punctuation
```

```python
def remove_punc0(text):
    for char in exclude:
        text = text.replace(char, '')
    return text
```

```python
text = 'string.with.punc?'
```

```python
start = time.time()
data['review'].apply(remove_punc0)
time1 = time.time() - start
time1
```

### Standard and Fast Method

```python
def remove_punc1(text):
    return text.translate(str.maketrans('', '',exclude))
```

```python
start = time.time()
data['review'].apply(remove_punc1)
time1 = time.time() - start
time1
```

## Stop Words removal

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
def remove_stopwords(text):
    return " ".join(
        [word for word in text.split() if word.lower() not in stop_words]
    )
```

```python
data['review'] = data['review'].apply(lambda x: remove_stopwords(x) if isinstance(x, str) else x)
```
